```python
# import necessary packages
import os
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
ethnicity = pd.read_csv("data/ethnicity17.csv",sep=";").rename(columns={'Kubus aantal inwoners (leeftijd, geslacht, migratieachtergrond)|Antilliaans|2017':'c_antilles',
       'Kubus aantal inwoners (leeftijd, geslacht, migratieachtergrond)|Autochtoon|2017':'c_dutch',
       'Kubus aantal inwoners (leeftijd, geslacht, migratieachtergrond)|Indonesisch|2017':'c_indonesian',
       'Kubus aantal inwoners (leeftijd, geslacht, migratieachtergrond)|Marokkaans|2017':'c_morocco',
       'Kubus aantal inwoners (leeftijd, geslacht, migratieachtergrond)|Oost-Europees|2017':'c_eastern_europe',
       'Kubus aantal inwoners (leeftijd, geslacht, migratieachtergrond)|Overig niet-Westers allochtoon|2017':'c_other_non_western',
       'Kubus aantal inwoners (leeftijd, geslacht, migratieachtergrond)|Overig Westers allochtoon|2017':'c_other_western',
       'Kubus aantal inwoners (leeftijd, geslacht, migratieachtergrond)|Surinaams|2017':'c_suriname',
       'Kubus aantal inwoners (leeftijd, geslacht, migratieachtergrond)|Turks|2017':'c_turkey'}).fillna(0).replace("x",0)
```

```python
ethnicity["Buurten"] = ethnicity["Buurten"].str.split()
```

```python
ethnicity["Buurten"] = [item[0] for item in ethnicity["Buurten"]]
```

```python
ethnicity = ethnicity.rename(columns={'Buurten':'Buurtcode'})
```

```python
ethnicity["Buurtcode"] = ethnicity["Buurtcode"].astype(int)
```

```python
ethnicity = ethnicity.set_index("Buurtcode")
```

```python
ethnicity.to_csv("output/ethnicity17.csv")
```
